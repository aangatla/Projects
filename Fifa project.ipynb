```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fifa = pd.read_csv("players_20.csv")
```

```python
fifa
```

```python
for col in fifa.columns:
    print(col)
```

```python
fifa.shape
# row in shape shows number of players and columns shows various details about players
```

```python
# To which nation most players in dataframe belong
nationality_players = fifa["nationality"].value_counts()
nationality_players
```

```python
top_nations = nationality_players[:5]
top_nations
```

```python
# plotting the bar plot showing top 5 nations to which players belong
nation_name = top_nations.index
plt.bar(nation_name, top_nations, color = "orange", width = 0.5)
plt.title('Top Football playing Nations',color = "green")
plt.xlabel("Nationality",color = "green")
plt.ylabel("Number of Players", color = "green")
```

```python
#Now we want to analyse salaries of players
player_salary_details = fifa[["short_name", "wage_eur"]]
player_salary_details.head()
```

```python
player_salary = player_salary_details.sort_values(by = "wage_eur",ascending = False)[:5]
# we have arranged player salary in decreasing order
```

```python
#now we will plot a bar plot of salary of top 5 players in terms of salary
x = player_salary["short_name"]
y = player_salary["wage_eur"]
plt.figure(figsize = (8,8))
plt.bar(x,y, color = "red", width = 0.5)
plt.title("Topmost Players Salary", color = "g")
plt.xlabel("Amount(in Euros)",color = "g")
plt.ylabel("Player name", color = "g")
```

```python
# Analysing German nationality player
filt = fifa["nationality"]== "Germany"
german_fifa = fifa[filt]
german_fifa  #this is our new dataframe comprising of all those players which belong to Germany
```

```python
#arranging german players as per their height
german_fifa.sort_values(by = "height_cm", ascending = False)[["short_name", "height_cm"]].head(10)
```

```python
# arranging german players in decreasing order of their weight
german_fifa.sort_values(by = "weight_kg", ascending = False)[["short_name", "weight_kg"]].head(10)
```

```python
# arranging german players in decreasing order of their wages
german_fifa.sort_values(by = "wage_eur", ascending = False)[["short_name", "wage_eur"]].head(10)
```

```python
# analysing shooting skills of all players
player_shooting = fifa[["short_name", "shooting","nationality", "club"]]
player_shooting.sort_values(by = "shooting", ascending = False).head(10)
```

```python
# which nationality has most number of best shooting players-->we do this by analysing over top 
# 50 shooters
top_50_shooter = player_shooting.sort_values(by = "shooting", ascending = False).head(50)
top_50_shooter["nationality"].value_counts()
# we can see no single team has disproportionate numbe
```

```python
# analysing defending skills of players
player_defend = fifa[["short_name", "defending","nationality", "club"]]
player_defend.sort_values(by = "defending", ascending = False).head(10)
```

```python
# checking to which nationality most defenders belong-->checking for top 50 defender
top_50_defenders = player_defend.sort_values(by = "defending", ascending = False).head(50)
top_50_defenders["nationality"].value_counts()
# here we can see Brazil has max number of defenders and have 10 out of 50 defenders
```
